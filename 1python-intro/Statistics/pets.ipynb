```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

sns.set_style('darkgrid')
plt.rcParams['figure.dpi'] = 120
```

```python
df = pd.read_excel('catsvdogs.xlsx')
```

```python
# 1. Traza un histograma del porcentaje de poseedores de perros.
var = 'Percentage of Dog Owners'
df[var].hist()
plt.xlabel(var)
plt.ylabel('Frequency')
plt.xlim(0,50)
```

```python
# 2. Traza un histograma del porcentaje de poseedores de gatos.
var = 'Percentage of Cat Owners'
df[var].hist()
plt.xlabel(var)
plt.ylabel('Frequency')
plt.xlim(0,50)
```

**3a. ¿Qué podemos ver al comparar estos dos histogramas? ¿Qué información nos da?**

Los perros son más populares que los gatos

```python
dft = df[['Mean Number of Dogs per household', 'Mean Number of Cats']].melt()
categories = {'Mean Number of Dogs per household':'Dogs', 'Mean Number of Cats':'Cats'}
dft.sort_values('variable', inplace=True)
dft.variable = dft.variable.map(categories)

sns.boxplot(data=dft, x='variable', y='value', notch=True)
plt.ylabel('Número promedio de mascotas')
plt.xlabel(None)
```

**5. ¿Qué podemos ver al comparar estos dos diagramas de caja? ¿Qué información nos da?**

A pesar de que los perros son más populares que los gatos, en hogares que si tienen gatos el número promedio de gatos por hogar es mayor que el de perros. Esto indica que usualmente los hogares tienen un mayor número de gatos.
