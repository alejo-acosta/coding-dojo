- Descripción: ejercicio fizzbuz

```python
def fizzbuzz(i):
  fizz = 'fizz' if i%3==0 else ''
  buzz = 'buzz' if i%5==0 else ''
  num = str(i) if fizz+buzz=='' else ''
  return num+fizz+buzz
```

```python
for i in range(1,30):
  print(fizzbuzz(i))
```
