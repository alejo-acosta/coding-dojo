```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df = pd.read_csv('./clase5/data-alcohol.csv')
```

```python
df.head()
```

```python
colores = ['#951818',"#B1A924", "#4374B3",'#00FF13']
sns.set_palette(colores)
sns.color_palette()
```

```python
sns.boxplot(data=df)
```

```python
sns.distplot(df.beer_servings, norm_hist=True)
```

```python
sns.histplot(df.beer_servings, stat='count', kde=True)
```

```python
sns.set()
sns.set_style('whitegrid')
```

```python
sns.lmplot(data=df, x='beer_servings', y='wine_servings', ci=99)
```

```python
sns.set()
sns.set_style('darkgrid')
```

```python
sns.lmplot(data=df, x='beer_servings', y='wine_servings', fit_reg=False)
```

```python
plt.figure(figsize=(20,4))
sns.set_style('ticks')
sns.swarmplot(data=df.sample(20), x='country', y='wine_servings')
```
